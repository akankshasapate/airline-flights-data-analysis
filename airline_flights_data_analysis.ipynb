VII. Showing all the Airlines with their Number of Flights in Horizontal Bar Graph

```python
new_db['airline'].value_counts(ascending=True).plot.barh( color = ['lightgreen', 'lightblue'])

plt.title("Airlines with frequencies")

plt.xlabel(" Number of flights")

plt.ylabel(" Airlines")

plt.show()
```

2. Show Bar Graphs representing the Departure Time & Arrival Time

```python
new_db.head()
```

I. Showing the Departure Time for the flights

```python
new_db['departure_time'].value_counts()
```

II. Showing the Departure Time & Arrival Time for the flights with their counts

```python
plt.figure(figsize = (16,4))

plt.subplot(1,2,1)

plt.bar( new_db['departure_time'].value_counts().index , new_db['departure_time'].value_counts().values, color = ['r', 'b'] )
plt.title("Departure Time")
plt.xlabel("D. Time")
plt.ylabel("D. Freq")

plt.subplot(1,2,2)

plt.bar( new_db['arrival_time'].value_counts().index, new_db['arrival_time'].value_counts().values, color = ['g', 'y'])
plt.title("Arrival Time")
plt.xlabel("A. Time")
plt.ylabel("A. Freq")

plt.show()
```

II. Showing the Source City of the flights

III. Showing the Source City & Destination City for the flights with their counts

```python
plt.figure( figsize= (16,4))

plt.subplot(1,2,1)

plt.barh( new_db['source_city'].value_counts().index , new_db['source_city'].value_counts().values, color = ['m', 'g'])
plt.title("Source Cities with No. of flights")
plt.ylabel("Cities")
plt.xlabel("No. of flights")

plt.subplot(1,2,2)

plt.barh( new_db['destination_city'].value_counts().index , new_db['destination_city'].value_counts().values, color = ['y', 'r'])
plt.title("Destination Cities with No. of flights")
plt.ylabel("Cities")
plt.xlabel("No. of flights")

plt.show()
```

4. Does price varies with airlines ?

```python
new_db.head()
```

I. Grouping the airlines and checking their mean price

```python
new_db.groupby('airline')['price'].mean()
```

III. Drawing a Categorical Plot showing the Mean Ticket Price for each Airline

```python
sns.catplot( x = 'airline', y = 'price', kind = 'bar', palette = 'rocket', data = new_db, hue = 'class')

plt.show()
```
